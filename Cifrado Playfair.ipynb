**Notaci√≥n y convenciones**

* Sea $A=\{A,B,\dots ,Z\}$ el alfabeto latino en may√∫sculas.  
* Como el cuadrado de Playfair es de dimensi√≥n $5\times5$, se fusionan $I$ y $J$ en un √∫nico s√≠mbolo $I/J$; as√≠ el alfabeto efectivo ser√° $\tilde A = A\setminus\{J\}$ con $|\tilde A|=25$.  
* Denotaremos por $\mathbb{Z}_5=\{0,1,2,3,4\}$ el anillo de enteros m√≥dulo¬†$5$.

---
**Generaci√≥n de la matriz clave**
1. Sea $K\in A^{\ast}$ (cadena de letras) la palabra clave.  
2. Formamos la secuencia  
   $S = \bigl(K\;\Vert\; \tilde A\bigr)$  
   (concatenaci√≥n de la clave seguida del alfabeto sin¬†$J$).  
3. Eliminar duplicados preservando la primera aparici√≥n ‚Üí obtenemos  
   $M = (m_0,m_1,\dots ,m_{24})$.  
4. Construimos la matriz  
$$
\mathbf{M}=
\begin{pmatrix}
m_0 & m_1 & m_2 & m_3 & m_4\\
m_5 & m_6 & m_7 & m_8 & m_9\\
m_{10}&m_{11}&m_{12}&m_{13}&m_{14}\\
m_{15}&m_{16}&m_{17}&m_{18}&m_{19}\\
m_{20}&m_{21}&m_{22}&m_{23}&m_{24}
\end{pmatrix}.
$$

---
Definimos la funci√≥n de localizaci√≥n  $\mathrm{pos}: \tilde A \to \mathbb{Z}_5\times\mathbb{Z}_5$  tal que $\mathrm{pos}(c)=(r,c)$ si $c$ est√° en la fila¬†$r$ y columna¬†$c$ de¬†$\mathbf{M}$.

---

***Pre‚Äëprocesado del texto claro***
* Sea $P\in A^{\ast}$ el texto plano.  
* Normalizamos: quitar acentos/puntuaci√≥n, convertir a may√∫sculas, reemplazar $J\to I$.  
* Dividimos $P$ en bigramas aplicando:  
  * Si un par $(x,x)$ contiene dos letras id√©nticas, se intercala un relleno $X$ formando $(x,X)$.  
  * Si el texto resultante tiene longitud impar, se agrega un $X$ final.

As√≠ obtenemos la lista ordenada de pares  
$\mathcal{P}= \bigl(p_1,p_2,\dots ,p_t\bigr), \qquad p_i=(a_i,b_i)\in\tilde A^2$.

---
**Funci√≥n de cifrado sobre un par**

Sea $(a,b)\in\tilde A^2$ con  $\mathrm{pos}(a)=(r_a,c_a)$ y $\mathrm{pos}(b)=(r_b,c_b)$.

La transformaci√≥n $\mathcal{E}:\tilde A^2\to\tilde A^2$ se define:

1. **Misma fila**  
   $r_a=r_b \; \Longrightarrow\;  
   \mathcal{E}(a,b)=\bigl( \mathbf{M}_{r_a,(c_a+1)\bmod5},\;\mathbf{M}_{r_b,(c_b+1)\bmod5}\bigr)$

2. **Misma columna**  
   $c_a=c_b \; \Longrightarrow\;  
   \mathcal{E}(a,b)=\bigl( \mathbf{M}_{(r_a+1)\bmod5,\,c_a},\;\mathbf{M}_{(r_b+1)\bmod5,\,c_b}\bigr)$
3. **Rect√°ngulo**  
   $ \mathcal{E}(a,b)=\bigl( \mathbf{M}_{r_a,c_b},\; \mathbf{M}_{r_b,c_a}\bigr)$

El texto cifrado completo es  
$C = \mathcal{E}(p_1)\Vert\mathcal{E}(p_2)\Vert\cdots\Vert\mathcal{E}(p_t)$.

---

**Descifrado**

La inversa $\mathcal{D}$ usa las mismas reglas con desplazamientos opuestos:

1. Fila: $(c_a-1)\bmod5$.  
2. Columna: $(r_a-1)\bmod5$.  
3. Rect√°ngulo: intercambio de columnas como en el cifrado.

**Propiedades algebraicas**

* $\mathcal{E}$ es una permutaci√≥n sobre $\tilde A^2$ ‚Üí es biyectiva.  

* El espacio de claves (todas las matrices posibles) tiene  
  $\dfrac{25!}{(25-|K|)!}$ configuraciones distintas (donde $|K|$ es la longitud de la clave sin repeticiones).

Se escoge la letra 'X' porque es poco frecuente en ingl√©s, de modo que el receptor reconoce f√°cilmente que es un relleno artificial y la descarta sin riesgo de confundirla con contenido real.

**üì•Importacionesüì¶**

```python

!pip install -q ipywidgets

import unicodedata, string, re
import ipywidgets as wd
from IPython.display import display
```

**üë®‚ÄçüíªImplementaci√≥nüë©‚Äçüíª**

```python

ALPHABET = "ABCDEFGHIKLMNOPQRSTUVWXYZ"
def _normalize(text: str) -> str:
    t = unicodedata.normalize("NFKD", text)
    t = "".join(c for c in t if not unicodedata.combining(c))
    t = re.sub(r"[^A-Z]", "", t.upper()).replace("J", "I")
    return t
```

**üë®‚ÄçüíªImplementaci√≥nüë©‚Äçüíª**

```python

def build_matrix(keyword: str):

    kw = _normalize(keyword)
    seen, seq = set(), []
    for ch in kw + ALPHABET:
        if ch not in seen:
            seq.append(ch); seen.add(ch)
    matrix = [seq[i:i+5] for i in range(0, 25, 5)]
    pos = {matrix[r][c]: (r, c) for r in range(5) for c in range(5)}
    return matrix, pos

def matrix_html(mat):
    rows = ["<table style='font-family:monospace;border-collapse:collapse;'>"]
    for r in mat:
        rows.append("<tr>" + "".join(f"<td style='border:1px solid #888;padding:4px'>{c}</td>" for c in r) + "</tr>")
    rows.append("</table>")
    return "".join(rows)
```

**üë®‚ÄçüíªImplementaci√≥nüë©‚Äçüíª**

```python

def prepare_pairs(text: str):

    t = _normalize(text)
    pairs = []
    i = 0
    while i < len(t):
        a = t[i]
        b = t[i+1] if i+1 < len(t) else "X"
        if a == b:
            pairs.append(a + "X")
            i += 1
        else:
            pairs.append(a + b)
            i += 2
    if len(pairs[-1]) == 1:
        pairs[-1] += "X"
    return pairs
```

**üë®‚ÄçüíªImplementaci√≥nüë©‚Äçüíª**

```python

def playfair(text: str, key: str, decrypt: bool = False) -> str:
    mat, pos = build_matrix(key)
    pairs = prepare_pairs(text) if not decrypt else [text[i:i+2] for i in range(0, len(text), 2)]
    result = []
    for p in pairs:
        a, b = p[0], p[1]
        ra, ca = pos[a]
        rb, cb = pos[b]
        if ra == rb:
            if decrypt:
                ca, cb = (ca - 1) % 5, (cb - 1) % 5
            else:
                ca, cb = (ca + 1) % 5, (cb + 1) % 5
        elif ca == cb:
            if decrypt:
                ra, rb = (ra - 1) % 5, (rb - 1) % 5
            else:
                ra, rb = (ra + 1) % 5, (rb + 1) % 5
        else:
            ca, cb = cb, ca
        result.append(mat[ra][ca] + mat[rb][cb])
    return "".join(result)
```

**üë®‚ÄçüíªImplementaci√≥nüë©‚Äçüíª**

```python

txt_input = wd.Textarea(value="LENGUAJE",description="Texto:",layout=wd.Layout(width="100%", height="80px"))
key_input = wd.Text(value="LENGUAJE",description="Clave:")
mode = wd.ToggleButtons(options=[("Cifrar", False), ("Descifrar", True)],description="Modo:")
output = wd.Output(layout=wd.Layout(border="1px solid #888", padding="6px"))
matrix_box = wd.HTML()

def refresh(*_):
    output.clear_output()
    matrix_box.value = ""
    with output:
        if not key_input.value.strip():
            print("‚ö†Ô∏è  La clave no puede estar vac√≠a."); return
        res = playfair(txt_input.value, key_input.value, decrypt=mode.value)
        if mode.value:
            print("üîì Texto descifrado:")
        else:
            print("üîê Texto cifrado:")
        print(res)

    matrix_box.value = matrix_html(build_matrix(key_input.value)[0])

for w in (txt_input, key_input, mode):
    w.observe(refresh, names="value")
```

**üë®‚ÄçüíªImplementaci√≥nüë©‚Äçüíª**

```python

display(wd.VBox([wd.HTML("<h3 style='margin:0'>Cifrado Playfair</h3>"),txt_input, key_input, mode,output,wd.HTML("<b>Matriz 5√ó5 generada:</b>"),matrix_box]))
refresh()
```

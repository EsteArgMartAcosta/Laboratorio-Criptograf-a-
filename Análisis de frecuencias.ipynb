```python

html = f"""
<style>
  body {{ font-family: 'Segoe UI', Tahoma, sans-serif; margin: 0; padding: 10px; }}
  h1 {{ text-align:center; color: #333; margin-bottom: 6px; }}
  textarea {{ width:100%; box-sizing:border-box; font-family:monospace; font-size:0.9em; }}
  .myButton {{ display:inline-block; margin:6px 4px; padding:6px 14px;
    background: linear-gradient(to bottom, #8e44ad, #663399);
    color:white; border:none; border-radius:4px; cursor:pointer; }}
  .myButton:hover {{ background: linear-gradient(to bottom, #663399, #4b0082); }}
  table {{ width:100%; border-collapse:collapse; margin-top:8px; }}
  .columnChart td {{ vertical-align:bottom; height:100px; text-align:center; }}
  .columnChart .column {{ display:block; width:100%; background:#8e44ad; border-radius:4px 4px 0 0; }}
  .chartPercentages td {{ text-align:center; font-size:0.8em; color:#555; padding-top:4px; }}
  .chartLabels td {{ text-align:center; font-weight:bold; padding-top:4px; }}
  .mapping td {{ padding:4px; text-align:center; }}
  .mapping input {{ width:1.2em; text-align:center; font-weight:bold; }}
  #plain {{ height:120px; margin-top:8px; }}
</style>

<h1>Romper Cifrado por Frecuencias</h1>

<label><strong>Texto cifrado:</strong></label><br>
<textarea id="cipher" rows="6" placeholder="Pega aqu√≠ el texto cifrado‚Ä¶"></textarea>
<br>
<button class="myButton" onclick="frequencyAnalysis()">1. Analizar Frecuencias</button>
<button class="myButton" onclick="breakByFrequency()">2. Romper por Frecuencia</button>

<table>
  <tr class="columnChart">
    {''.join(f'<td id="bar_{ch}"></td>' for ch in string.ascii_lowercase)}
  </tr>
  <tr class="chartPercentages">
    {''.join(f'<td id="pct_{ch}"></td>' for ch in string.ascii_lowercase)}
  </tr>
  <tr class="chartLabels">
    {''.join(f'<td>{ch.upper()}</td>' for ch in string.ascii_lowercase)}
  </tr>
  <tr class="mapping">
    {''.join(f'<td><input id="map_{ch}" maxlength="1" placeholder="_"></td>' for ch in string.ascii_lowercase)}
  </tr>
</table>

<label><strong>Texto descifrado:</strong></label><br>
<textarea id="plain" rows="6" readonly placeholder="Aqu√≠ aparece el texto descifrado‚Ä¶"></textarea>

<script>
function normalize(s) {{
  s = s.normalize('NFD').replace(/[\u0300-\u036f]/g,'');
  return s.toUpperCase().replace(/[^A-Z]/g,'');
}}

function frequencyAnalysis() {{
  const text = normalize(document.getElementById('cipher').value);
  const freq = {{}}; 'abcdefghijklmnopqrstuvwxyz'.split('').forEach(l=>freq[l]=0);
  for(let c of text) if(c.toLowerCase() in freq) freq[c.toLowerCase()]++;
  const maxc = Math.max(...Object.values(freq),1);
  const scale = 100/maxc;
  const total = text.length||1;
  for(let l of 'abcdefghijklmnopqrstuvwxyz') {{
    const c = freq[l], h = c*scale;
    document.getElementById('bar_'+l).innerHTML = c>0
      ? '<div class="column" style="height:'+h+'px;"></div>' : '';
    document.getElementById('pct_'+l).textContent = c>0
      ? ((c/total)*100).toFixed(1)+'%' : '';
  }}
}}

function breakByFrequency() {{
  // Orden de frecuencia aproximado en espa√±ol: EASRNDLCTUMCOPBIQFHVGYXZJ√ë√ú
  const ref = 'EASRNDLCTUMCOPBIQFHVGYXZJ√ë√ú'.toLowerCase().split('');
  // Ordenar letras del texto cifrado por frecuencia
  const freq = {{}}; 'abcdefghijklmnopqrstuvwxyz'.split('').forEach(l=>freq[l]=0);
  const text = normalize(document.getElementById('cipher').value);
  for(let c of text) if(c.toLowerCase() in freq) freq[c.toLowerCase()]++;
  const sorted = Object.entries(freq)
    .sort((a,b)=>b[1]-a[1])
    .map(e=>e[0]);
  // Asignar mapeo autom√°tico
  for(let i=0;i<sorted.length;i++) {{
    const src = sorted[i], tgt = ref[i]||'_';
    document.getElementById('map_'+src).value = tgt.toUpperCase();
  }}
  applyMapping();
}}

function applyMapping() {{
  const text = document.getElementById('cipher').value;
  const map = {{}};
  'abcdefghijklmnopqrstuvwxyz'.split('').forEach(l=>{{
    const v = document.getElementById('map_'+l).value.toUpperCase();
    if(v.match(/[A-Z√ë√ú]/)) map[l] = v;
  }});
  let out = '';
  for(let ch of text) {{
    const c = ch.toLowerCase();
    if(c in map) {{
      out += (ch === ch.toUpperCase() ? map[c] : map[c].toLowerCase());
    }} else {{
      out += ch;
    }}
  }}
  document.getElementById('plain').value = out;
}}

// Permitir que el usuario modifique cualquier casilla y ver el resultado en vivo
'abcdefghijklmnopqrstuvwxyz'.split('').forEach(l=>{{
  document.addEventListener('input', function(e) {{
    if(e.target && e.target.id === 'map_'+l) applyMapping();
  }});
}});

</script>
"""
```
